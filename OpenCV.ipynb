```python
plt.imshow(org_img)
```

```python
gray_img = cv2.imread('sample_img.jpeg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(gray_img)
```

```python
gray_img2 = cv2.cvtColor(org_img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_img2)
```

```python
resize_img = cv2.resize(org_img,(1000,600))
```

```python
plt.imshow(resize_img)
```

```python
#resize using ratio
w_ratio = 0.5
h_ratio = 0.5
```

```python
resize_img_ratio = cv2.resize(org_img,(0,0),org_img,w_ratio,h_ratio)
```

```python
plt.imshow(resize_img_ratio)
```

```python
img2 = cv2.imread('sample_flip.jpeg')
```

```python
plt.imshow(img2)
```

```python
flip_img = cv2.flip(img2,0) #flip on horizontal axis
```

```python
plt.imshow(flip_img)
```

```python
flip_img_y = cv2.flip(img2,1)
```

```python
plt.imshow(flip_img_y)
```

```python
cv2.imwrite('sample_imwrite.jpeg',flip_img_y)
```
