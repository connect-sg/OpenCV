```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('samplesnookerball.jpeg')
```

```python
def display(img):
     fig = plt.figure(figsize=(10,10))
     ax = fig.add_subplot(111)
     ax.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
display(img)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
kernel = np.full((5,5),1,dtype=np.int8)
```

```python
morph_img = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
```

```python
display(morph_img)
```

```python
erode_img = cv2.erode(morph_img,kernel,iterations=4)
```

```python
display(erode_img)
```

```python
ret,thres = cv2.threshold(erode_img,127,255,cv2.THRESH_TRUNC)
```

```python
display(thres)
```
