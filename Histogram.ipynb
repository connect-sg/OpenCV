```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('sample_img.jpeg')
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
color = ('R','G','B')

for i,col in enumerate(color):
    hist = cv2.calcHist(img,[i],None,[256],[0,256])
    plt.plot(hist,color = col)
    plt.xlim([0,256])
    
```
