```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
tiger = cv2.imread('tiger-siberian.jpeg')
```

```python
type(tiger)
```

```python
savetiger = cv2.imread('savetiger.jpeg')
```

```python
type(savetiger)
```

```python
tiger = cv2.cvtColor(tiger,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(tiger)
```

```python
savetiger = cv2.cvtColor(savetiger,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(savetiger)
```

```python
savetiger.shape
```

```python
tiger_resize = cv2.resize(tiger,(750,1000))
```

```python
tiger_resize.shape
```

```python
savetiger_resize = cv2.resize(savetiger,(750,1000))
```

```python
savetiger_resize.shape
```

```python
blended_img = cv2.addWeighted(src1=tiger_resize, alpha =0.8,src2=savetiger_resize, beta =0.3,gamma =0)
```

```python
plt.imshow(blended_img)
```

```python
savetiger = cv2.resize(savetiger,(650,650))
```

```python
savetiger.shape
```

```python
plt.imshow(savetiger)
```

```python
savetigergray = cv2.cvtColor(savetiger,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(savetigergray,cmap='gray')
```

```python
savetigergray.shape
```

```python
mask_inv = cv2.bitwise_not(savetigergray)
```

```python
mask_inv.shape
```

```python
plt.imshow(mask_inv,cmap='gray')
```

```python
white_img = np.full(savetiger.shape,255,dtype =np.uint8)
```

```python
white_img.shape
```

```python
bkgrnd = cv2.bitwise_or(white_img,white_img,mask=mask_inv)
```

```python
plt.imshow(bkgrnd)
```

```python
img = cv2.bitwise_or(savetiger,savetiger,mask=mask_inv)
```

```python
plt.imshow(img)
```

```python
image = tiger[0:650,0:650]
```

```python
plt.imshow(image)
```

```python
savetiger.shape
```

```python
image_blend = cv2.bitwise_or(image,img)
```

```python
plt.imshow(image_blend)
```

```python
tiger[0:650,0:650]=image_blend
```
