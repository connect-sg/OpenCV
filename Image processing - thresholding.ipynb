```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('colorful.jpeg')
```

```python
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
img.max()
```

```python
ret,thres1 = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(thres1,cmap='gray')
```

```python
ret,thres2 = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thres2,cmap='gray')
```

```python
plt.imshow(thres3,cmap='gray')
```

```python
plt.imshow(thres4,cmap='gray')
```
